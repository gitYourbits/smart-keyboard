```python
with open('/content/dataset_book.txt', 'r') as f:
  text = f.read()
type(text)
```

```python
txt = ''
for i in text:
  if i=='\n':
    txt+=' '
  else:
    txt+=i


# textList = text.split('\n')
# text = ''
# for i in textList:
#   text+=i

text = txt.split('.')
```

```python
uniques = []
for i in text:
  if i not in uniques:
    uniques.append(i)

uniques
```

```python
spaceCount=0
for i in uniques:
  for j in i:
    if j == ' ':
      spaceCount+=1
  if spaceCount<3 or len(i)<15:
    uniques.remove(i)

```

```python
# x = min(len(i) for i in uniques)
# print(uniques)
uList=[]
for i in uniques:
  uList.append(i.split(' '))
  # print(type(i))
print(uList[0])
```

```python
def isValidWord(word):
  for i in word:
    if i not in 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ':
      return False
  return True
```

```python
for i in uList:
  while '' in i:
    i.remove('')
  for j in i:
    if not isValidWord(j):
      i.remove(j)
uList[0]
```

```python
for i in uList:
  if len(i)<3:
    print(i)
    uList.remove(i)
# uList
```

```python
for i in uList:
  if len(i)<3:
    uList.remove(i)
```

```python
sentences=[' '.join(i) for i in uList]
text='\n'.join(sentences)
```

```python
import pandas as pd
df = pd.DataFrame(sentences, columns=["message"])
df.head()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['message'])
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index)
vocab_size
```

```python

X = []
for message in df['message']:
  for i in range(1, len(tokenizer.texts_to_sequences([message])[0])):
    X.append(tokenizer.texts_to_sequences([message])[0][:i+1])
```

```python
X[:10]
```

```python
longest_sen = max(len(i) for i in X)
longest_sen
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X = pad_sequences(X, maxlen=longest_sen, padding='pre')
```

```python
y=X
X=X[:,:-1]
y=y[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=vocab_size+1)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
embedding_dim = 50
sequence_length = 107
```

```python
model = Sequential()
model.add(Embedding(vocab_size+1, embedding_dim))
model.add(LSTM(256))
model.add(Dense(vocab_size+1, activation='softmax'))

# Build the model explicitly with the input shape
model.build(input_shape=(None, sequence_length))  # Here, None is the placeholder for any batch size
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
